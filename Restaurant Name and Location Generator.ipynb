```python
import math

def haversine_distance(lat1, lat2):
    # Convert latitude values from degrees to radians
    lat1, lat2 = map(math.radians, [lat1, lat2])

    # Radius of the Earth in kilometers
    R = 6371

    # Haversine formula for latitude values
    dlat = lat2 - lat1
    a = math.sin(dlat/2)**2
    c = 2 * math.asin(math.sqrt(a))

    # Calculate the distance between two latitude values in kilometers
    lat_distance = R * c

    return lat_distance

def longitude_distance(lon1, lon2):
    # Convert longitude values from degrees to radians
    lon1, lon2 = map(math.radians, [lon1, lon2])

    # Radius of the Earth in kilometers
    R = 6371

    # Calculate the distance between two longitude values in kilometers
    lon_distance = R * math.acos(math.cos(lon1) * math.cos(lon2) * math.cos(0) + math.sin(lon1) * math.sin(lon2))

    return lon_distance
```
